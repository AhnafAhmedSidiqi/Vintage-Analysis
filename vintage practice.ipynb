```python
cols = df.columns.tolist()

num_cols_to_rename = len(cols) - 3
new_names = [f'M{i+1}' for i in range(num_cols_to_rename)]

new_column_names = cols[:3] + new_names

df.columns = new_column_names

print(df.columns)
             
```

#### def left_align_row(row):
    # Keep only non-null values (NaN, None, pd.NA)
    non_null = row.dropna().values
    # Create an object array filled with NaN (as object)
    aligned = np.full(len(row), np.nan, dtype=object)
    # Place non-null values at the beginning
    aligned[:len(non_null)] = non_null
    return pd.Series(aligned, index=row.index)

df1 = df.apply(left_align_row, axis=1)

```python
df.head()
```

```python
cols_to_update = df1.columns[3:]

df[cols_to_update] = df[cols_to_update].apply(lambda X:np.where( X.isna(), np.nan, np.where(X > 4,1,0)))
```

```python
df[cols_to_update]
```

```python
def propagate_until_blank(row):
    # Initialize `activate` to a default value before the loop
    activate = False
    new_row = row.copy()
    for i in range(len(new_row)):
        if new_row[i] == 1:
            activate = True
        elif activate:
            new_row[i] = 1
    return new_row

df[cols_to_update]= df[cols_to_update].apply(propagate_until_blank, axis = 1)
    
```

```python
# Calculate the sum for the specified columns
sum_row = df[cols_to_update].sum()

# Print the result
print("\nSum row for selected columns:")
print(sum_row)
```

```python
# Calculate the count of non-null values for the specified columns
count_row = df[cols_to_update].count()

# Print the result
print("\nCount row (non-null values) for selected columns:")
print(count_row)
```

```python
# Create new Series for the summary rows, including an index label
sum_row_series = pd.Series(df[cols_to_update].sum(), name='Sum')
count_row_series = pd.Series(df[cols_to_update].count(), name='Count')

# Combine the summary rows into a new DataFrame
summary_df = pd.DataFrame([sum_row_series, count_row_series])

# Append the summary rows to the original DataFrame
df_with_summary = pd.concat([df[cols_to_update], summary_df])

print("\nDataFrame with summary rows:")
print(df_with_summary)
```

```python
# Transpose the summary_df
transposed_summary_df = summary_df.T 
Bad = transposed_summary_df['Sum']
Count = transposed_summary_df['Count']
```

```python
# 5. Plot the transposed summary DataFrame as a line chart
transposed_summary_df.plot(kind='line', figsize=(10, 6), marker='o')

# 6. Set the title and axis labels using plt
plt.title('Sum and Count per Column (Line Plot)')
plt.ylabel('Value') # This sets the y-label for the line chart
plt.xlabel('Columns')

# 7. Add a grid and legend for clarity
plt.grid(axis='Y', linestyle='--', alpha=0.7)
plt.legend(title='Metric')

# 8. Display the plot
plt.show()
```

```python
# Create a figure with two subplots side-by-side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# --- Left Subplot (Bar Chart) ---
transposed_summary_df['Sum'].plot(kind='bar', ax=ax1, color='skyblue')
ax1.set_title('Sum per Column')
ax1.set_xlabel('Columns')
ax1.set_ylabel('Sum Value')
ax1.tick_params(axis='x', rotation=0)
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# --- Right Subplot (Line Chart) ---
transposed_summary_df['Count'].plot(kind='line', ax=ax2, marker='o', color='lightgreen')
ax2.set_title('Count per Column')
ax2.set_xlabel('Columns')
ax2.set_ylabel('Count of non-null values')
ax2.tick_params(axis='x', rotation=0)
ax2.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
